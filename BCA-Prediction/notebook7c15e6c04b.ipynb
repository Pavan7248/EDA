# 4. Visualization

* Number of matches played in each IPL season 

```python
plt.figure(figsize= (15,7))
sns.countplot(x='Season',
             data=df_matches,
             order = df_matches['Season'].value_counts().index.sort_values())
plt.xlabel('Season',fontsize=12)
plt.ylabel('No: of matches',fontsize=12)
plt.title('Number of matches played in each IPL season',fontsize=16)
```

This graph illustrates that the highest number of matches were played in IPL-2013 

 * Matches won by Teams

```python
plt.figure(figsize = (11,9))
sns.countplot(y = 'winner',
              data = df_matches,
              order= df_matches['winner'].value_counts().index,
              palette = 'copper' )
plt.xlabel('No: of Wins',fontsize=12)
plt.ylabel('Team',fontsize=12)
plt.title('Matches won by the Teams',fontsize=16)
```

The above graph demonstrate that Mumbai indians won the highest number of matches in all seasons ,this states that mumbai indians is the strongest team. 

* Number of matches played in stadium

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'venue',
              data = df_matches,
              order = df_matches['venue'].value_counts().iloc[:10].index,
              palette='gist_heat')
plt.xlabel('No: of matches',fontsize=12)
plt.ylabel('Venue',fontsize=12)
```

Most of the matches were playes in Eden Gardens

* Toss Decision

* Toss Decision to field or bat in each ipl season

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season',
              hue='toss_decision',
              data=df_matches,
              order = df_matches['Season'].value_counts().index.sort_values(),
              palette='deep')
plt.xlabel('Season',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Decision to field or bat in each IPL season', fontsize=16)
```

* Player of the match

```python
plt.figure(figsize=(15,7))
player_of_match=df_matches['player_of_match'].value_counts()[:10]
sns.barplot(x = player_of_match.index,
            y = player_of_match.values,
            palette='cividis')

plt.title("Player of the Match",fontsize=15)
plt.xlabel('Player',fontsize=12)
plt.ylabel('Count',fontsize=12)
```

Above illustration states that CH Gayle was the Player of the match.
